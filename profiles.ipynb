```python
import netCDF4 as nc
import numpy as np
import torch
import matplotlib.pyplot as plt
```

```python
path = "data/6901766/MR6901766_011.nc"
ds = nc.Dataset(path)
```

```python
vars=[]
for var in ds.variables:
    vars.append(var)  
```

```python
def plot_profile(element):
    if element not in vars:
        print('measurement not avaiable')
        return
    if element == 'NITRATE':
        nitrate = ds['NITRATE'][:].data
        pres_nitrate = - ds['PRES_NITRATE'][:].data
        plt.style.use('dark_background')
        fig, ax = plt.subplots()
        ax.plot(nitrate, pres_nitrate, color = 'r')
        ax.scatter(nitrate, pres_nitrate, label = 'nitrate', marker = '.', color = 'r')
        ax.set(xlabel="nitrate values", ylabel="pressure")
        ax.xaxis.tick_top()
        plt.legend()
    if element == 'CHLA':
        chla = ds['CHLA'][:].data
        pres_chla = - ds['PRES_CHLA'][:].data
        plt.style.use('dark_background')
        fig, ax = plt.subplots()
        ax.plot(chla, pres_chla, color = 'r')
        ax.scatter(chla, pres_chla, label = 'chla', marker = '.', color = 'r')
        ax.set(xlabel="chla values", ylabel="pressure")
        ax.xaxis.tick_top()
        plt.legend()
```

```python
plot_profile('NITRATE')
```
